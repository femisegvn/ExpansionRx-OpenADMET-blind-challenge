# Exploratory Data Analysis

Figures to include:
- Tree plot of the training set and the test set
- Probability Distributions of all the endpoints


```python
from data.test_data import test_df
from data.train_data import train_df
from data.utils import characterise
import pandas as pd

test_df = characterise(test_df, name = 'Test')
train_df = characterise(train_df, name = 'Train')

df = pd.concat([train_df, test_df])
df
```

```python
import seaborn as sns
%config InlineBackend.figure_format = 'svg'
prop_pairs = [('MWt', 'TPSA'), ('MWt', 'CLogP')]#, ('CLogP', '')]

for pair in prop_pairs:
    sns.jointplot(data=df, x=pair[0], y=pair[1], ratio=3, hue='Dataset', palette = ["#7920d3","#e69f00"])
```

```python
from data.utils import get_train_test_similarity

df = get_train_test_similarity(df)
df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
sns.histplot(
    df.query("Dataset == 'Train'"), x="train_test_sim", color="#7920d3",
    stat="probability", common_norm=False,bins=20,label="Potency",alpha=0.8, element="bars",edgecolor="white")
plt.xlabel("Tanimoto similarity between train and test sets")
plt.ylabel("Probability")
plt.show()
```
